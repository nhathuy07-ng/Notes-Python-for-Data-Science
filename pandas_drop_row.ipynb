## Pandas: Drop Rows or Columns from DataFrame

Use `df.drop(labels, axis)`:
- labels: list of column or row indices
- axis: 1 if dropping columns, 0 if dropping rows

```python
import pandas as pd
df = pd.read_csv("datasets/players_20.csv", encoding="unicode-escape")
df.set_index("short_name", inplace=True)
df
```

### Drop Row

```python
# Drop one row using axis=0
df.drop("L. Messi", axis=0)
```

```python
# Drop one row using index parameters
df.drop(index=['L. Messi']) # no need for axis parameter. Only remove index rows.
```

```python
# Drop two or more rows and update DF (inplacce = True)
df.drop(index=['L. Messi', 'Cristiano Ronaldo'], inplace=True)
df.head(5)
```

### Drop Columns

```python
# Drop one column

# axis = 1
df.drop("long_name", axis=1)
```

```python
# Drop last cols
print(df.columns[[-1]])
df.drop(df.columns[[-1]], axis=1)
```

```python
# Drop two or more cols
df.drop(["long_name", "dob", "height_cm", "weight_kg"], axis=1)
```

```python
# Drop a range of rows
df.drop(df.index[1:9]) # drop rows 1 to 8
```

```python
# Drop a range of cols
df.drop(df.columns[0:9], axis=1) # drop cols 0 to 8
```

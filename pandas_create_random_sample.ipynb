## Pandas: Creating random sample using `sample()`

`sample()` extracts a random sample from DataFrames and Series

```python
import pandas as pd

df = pd.read_csv("datasets/players_20.csv", encoding='unicode-escape')
df.set_index("short_name", inplace=True)
df
```

```python
# Extract 10 random elements from the nationality column

df['nationality'].sample(10)
```

```python
# random_state locks the random result to the same set every run.
df['nationality'].sample(10, random_state=36)
```

```python
# Extract 20% sample of the dataframe

df.sample(frac=0.2)
```

```python
# Verify
18278/5
```

### upsample: increase the sampling rate with `frac > 1`:

**NOTE**: We must set **`replace = True`**. This is because, `replace = False` forces Pandas to use each data point only once, and upsamping requires reusing some of the data points.

```python
df2 = df.sample(frac=2, replace=True)
```

```python
df2
```

```python
# verify that upsampling reuses the already available data points
df2.value_counts("short_name")
```

### Filter a DataFrame with the `query()` method

- `query()`: works like filtering by boolean slicing, but more concise.
- Syntax: `df.query("condition")`
- Ex: `df.query("overall > 90")` same as `df[df.overall > 90]`
- Uses `&`, `|` and `~` (not) just like boolean slicing.
- Extra: use `.astype("new_type")` to convert type of a Series in Pandas

=> Think of it like SQL `WHERE`

```python
import pandas as pd

df = pd.read_csv("datasets/players_20.csv")
df.set_index("short_name", inplace=True)
```

```python
df
```

```python
# Select players older than 34

df.query("age > 34").age
```

```python
# same as boolean slicing
df[df.age > 34].age
```

```python
df.query("age>34 & overall > 83")[["age", "overall"]]
```

### Using & operator

```python
df.query("(overall>=85) & (nationality=='Argentina')")[["overall", "nationality"]]
```

```python
df.query("age>34 & nationality=='Italy'")
```

```python
# Same as...
df[(df.age > 34) & (df.nationality=='Italy')]
```

### Using ~ operator

```python
df.query("~(overall>=50)").overall
```

```python
# Select players born after 1990. Check out dtypes. Convert "dob" to datetime type.

# check datatype of DOB
df.dob.dtypes
```

```python
# query
_ = df.query("dob >= '1991-01-01'").sort_values('dob').dob
# or...
df.query("dob.dt.year > 1990").sort_values('dob').dob
```
